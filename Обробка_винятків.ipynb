–£ –≤–∏–ø–∞–¥–∫—É –Ω–∞—Å—Ç–∞–Ω–Ω—è –≤–∏–Ω—è—Ç–∫—É –æ—Ç—Ä–∏–º—É—î–º–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ —Ü—é –ø–æ–º–∏–ª–∫—É –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–Ω—è–º —Ç–∏–ø—É –ø–æ–º–∏–ª–∫–∏. –í –¥–∞–Ω–æ–º—É –≤–∏–ø–∞–¥–∫—É –º–∞—î–º–æ –ø–æ–º–∏–ª–∫—É `ZeroDivisionError` - –ø–æ–º–∏–ª–∫–∞ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å.

## –í–±—É–¥–æ–≤–∞–Ω—ñ –≤–∏–Ω—è—Ç–∫–∏

–£ Python —ñ—Å–Ω—É—î –±–∞–≥–∞—Ç–æ [**–≤–±—É–¥–æ–≤–∞–Ω–∏—Ö –≤–∏–Ω—è—Ç–∫—ñ–≤**](https://docs.python.org/uk/3.13/library/exceptions.html), —è–∫—ñ –≥–µ–Ω–µ—Ä—É—é—Ç—å—Å—è –ø—Ä–∏ –≤–∏–Ω–∏–∫–Ω–µ–Ω–Ω—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏—Ö –ø–æ–º–∏–ª–æ–∫. –û—Ç—Ä–∏–º–∞—Ç–∏ —Å–ø–∏—Å–æ–∫ —ó—Ö —É—Å—ñ—Ö —Ä–∞–∑–æ–º –º–æ–∂–Ω–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ—ó `locals()` —Ç–∞–∫–∏–º —á–∏–Ω–æ–º: `print(dir(locals()['__builtins__']))`

`ArithmeticError`, `AssertionError`, `AttributeError`, `BaseException`, `BlockingIOError`, `BrokenPipeError`, `BufferError`, `BytesWarning`, `ChildProcessError`, `ConnectionAbortedError`, `ConnectionError`, `ConnectionRefusedError`, `ConnectionResetError`, `DeprecationWarning`, `EOFError`, `EncodingWarning`, `EnvironmentError`, `FileExistsError`, `FileNotFoundError`, `FloatingPointError`, `IOError`, `ImportError`, `IndentationError`, `IndexError`, `InterruptedError`, `IsADirectoryError`, `KeyError`, `LookupError`, `MemoryError`, `ModuleNotFoundError`, `NameError`, `NotADirectoryError`, `NotImplementedError`, `OSError`, `OverflowError`, `PermissionError`, `ProcessLookupError`, `RecursionError`, `ReferenceError`, `RuntimeError`, `SyntaxError`, `SystemError`, `TabError`, `TimeoutError`, `TypeError`, `UnboundLocalError`, `UnicodeDecodeError`, `UnicodeEncodeError`, `UnicodeError`, `UnicodeTranslateError`, `ValueError`, `WindowsError`, `ZeroDivisionError`


–û—Å—å –¥–µ—è–∫—ñ –ø–æ—à–∏—Ä–µ–Ω—ñ –≤–∏–Ω—è—Ç–∫–∏ —ñ —ó—Ö –∑–Ω–∞—á–µ–Ω–Ω—è:

| –í–∏–Ω—è—Ç–æ–∫ | –ü—Ä–∏—á–∏–Ω–∞ –≤–∏–Ω–∏–∫–Ω–µ–Ω–Ω—è |
| :--- | :--- |
| `ImportError` | –ù–µ –∑–Ω–∞–π–¥–µ–Ω–æ —ñ–º–ø–æ—Ä—Ç–æ–≤–∞–Ω–∏–π –º–æ–¥—É–ª—å |
| `IndexError` | –ó–∞–∑–Ω–∞—á–µ–Ω–∏–π —ñ–Ω–¥–µ–∫—Å —É –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ –≤–∏—Ö–æ–¥–∏—Ç—å –∑–∞ –º–µ–∂—ñ –¥–æ–ø—É—Å—Ç–∏–º–æ–≥–æ –¥—ñ–∞–ø–∞–∑–æ–Ω—É |

| `OSError` | –ü–æ–º–∏–ª–∫–∞, –ø–æ–≤‚Äô—è–∑–∞–Ω–∞ —ñ–∑ –æ–ø–µ—Ä–∞—Ü—ñ–π–Ω–æ—é —Å–∏—Å—Ç–µ–º–æ—é |
| `OverflowError` | –†–µ–∑—É–ª—å—Ç–∞—Ç –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–æ—ó –æ–ø–µ—Ä–∞—Ü—ñ—ó –¥—É–∂–µ –≤–µ–ª–∏–∫–∏–π –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è |

| `UnboundLocalError` | –£ —Ñ—É–Ω–∫—Ü—ñ—ó –∞–±–æ –º–µ—Ç–æ–¥—ñ —Ä–æ–±–∏—Ç—å—Å—è –ø–æ—Å–∏–ª–∞–Ω–Ω—è –Ω–∞ –ª–æ–∫–∞–ª—å–Ω—É –∑–º—ñ–Ω–Ω—É, –∞–ª–µ –∑–Ω–∞—á–µ–Ω–Ω—è –Ω–µ –±—É–ª–æ –ø—Ä–∏–≤‚Äô—è–∑–∞–Ω–µ –¥–æ —Ü—ñ—î—ó –∑–º—ñ–Ω–Ω–æ—ó |

| `ValueError` | –§—É–Ω–∫—Ü—ñ—è –æ—Ç—Ä–∏–º—É—î –∞—Ä–≥—É–º–µ–Ω—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ —Ç–∏–ø—É, –∞–ª–µ –Ω–µ–∫–æ—Ä–µ–∫—Ç–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è —Ü—å–æ–≥–æ —Ç–∏–ø—É |

–ó–Ω–∞–Ω–Ω—è –≤–∏–Ω—è—Ç–∫—ñ–≤ –¥–æ–∑–≤–æ–ª—è—î –ø—Ä–æ–≥—Ä–∞–º–Ω–æ "—É–ø—ñ–π–º–∞—Ç–∏" –ø–æ–º–∏–ª–∫—É –≤ –∫–æ–¥—ñ –π –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –æ–±—Ä–æ–±–∏—Ç–∏.

## –û–±—Ä–æ–±–∫–∞ –≤–∏–Ω—è—Ç–∫—ñ–≤

–î–ª—è –æ–±—Ä–æ–±–∫–∏ –≤–∏–Ω—è—Ç–∫—ñ–≤ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –±–ª–æ–∫ `try...except` (*try* - "—Å–ø—Ä–æ–±—É–≤–∞—Ç–∏", *except* - "–∫—Ä—ñ–º", "–∑–∞ –≤–∏–Ω—è—Ç–∫–æ–º"). –í–∏–≥–ª—è–¥–∞—î —Ü–µ —Ç–∞–∫:

````python
try:
    # –ö–æ–¥, —è–∫–∏–π –º–æ–∂–µ —Å–ø—Ä–∏—á–∏–Ω–∏—Ç–∏ –≤–∏–Ω—è—Ç–æ–∫
except:
    # –ö–æ–¥, —è–∫–∏–π –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è —É —Ä–∞–∑—ñ –Ω–∞—Å—Ç–∞–Ω–Ω—è –≤–∏–Ω—è—Ç–∫—É
````

–ö–æ–ª–∏ –≤ –±–ª–æ–∫—É `try` –≤–∏–Ω–∏–∫–∞—î –≤–∏–Ω—è—Ç–æ–∫, –≤—ñ–Ω "–ø–µ—Ä–µ—Ö–æ–ø–ª—é—î—Ç—å—Å—è" –±–ª–æ–∫–æ–º `except` —ñ –ø—Ä–æ–≥—Ä–∞–º–∞ –∑–∞–≤–µ—Ä—à—É—î—Ç—å—Å—è —Ç–∏–º, —â–æ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–æ —É –Ω—å–æ–º—É.

**–ë–ª–æ–∫ `try` –º–æ–∂–µ —ñ—Å–Ω—É–≤–∞—Ç–∏ –±–µ–∑ –±–ª–æ–∫—É `except`, –∞ –æ—Ç –Ω–∞–≤–ø–∞–∫–∏ (`except` –±–µ–∑ `try`) - –Ω—ñ!**

–û—Å—å, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —è–∫–∏–π –≤–∏–≥–ª—è–¥ –º–∞—Ç–∏–º–µ –æ–±—Ä–æ–±–Ω–∏–∫ –≤ —Ä–∞–∑—ñ —Ä–∏–∑–∏–∫—É –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0:

````python
def division(x):
    try:
        y = 1/x
        return y
    except:
        return '–ù–∞ –Ω—É–ª—å –¥—ñ–ª–∏—Ç–∏ –Ω–µ –º–æ–∂–Ω–∞!'
print(division(5))
print(division(0))
````

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, —É —Ä–∞–∑—ñ –≤–∏–Ω–∏–∫–Ω–µ–Ω–Ω—è –ø–æ–º–∏–ª–∫–∏ –≤ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è—Ö, –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –æ—Ç—Ä–∏–º—É—î –Ω–µ –º–∞–ª–æ–∑—Ä–æ–∑—É–º—ñ–ª–µ –ª—é–¥–∏–Ω—ñ, —è–∫–∞ –Ω–µ –∑–Ω–∞—î Python, –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è `ZeroDivisionError`, –∞ —Ü—ñ–ª–∫–æ–º "–¥—Ä—É–∂–Ω—î" –¥–ª—è –Ω—å–æ–≥–æ –ø–æ—è—Å–Ω–µ–Ω–Ω—è —Ç–æ–≥–æ, —â–æ —Å—Ç–∞–ª–æ—Å—è.

## –û–±—Ä–æ–±–∫–∞ –≤–∏–Ω—è—Ç–∫—ñ–≤, –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —ó—Ö —Å–ø–µ—Ü–∏—Ñ—ñ–∫–∏

–ú–æ–≤–∞ –ø—Ä–æ —Ç–µ, —è–∫ –Ω–∞–≤—á–∏—Ç–∏ –ø—Ä–æ–≥—Ä–∞–º—É —Ä–µ–∞–≥—É–≤–∞—Ç–∏ –Ω–µ –Ω–∞ —è–∫—É—Å—å –ø–æ–º–∏–ª–∫—É –≤–∑–∞–≥–∞–ª—ñ, –∞ –Ω–∞ —Ü—ñ–ª–∫–æ–º –∫–æ–Ω–∫—Ä–µ—Ç–Ω—É, –∑–∞—Ä–∞–Ω—ñ –≤–∏–∑–Ω–∞—á–µ–Ω—É –ø—Ä–æ–≥—Ä–∞–º—ñ—Å—Ç–æ–º, –∞–¥–∂–µ –≤ —Ö–æ–¥—ñ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–¥—É –≤ –ø–µ–≤–Ω—ñ –º–æ–º–µ–Ω—Ç–∏ –º–æ–∂—É—Ç—å –≤–∏–Ω–∏–∫–∞—Ç–∏ —Ä—ñ–∑–Ω—ñ –ø–æ–º–∏–ª–∫–∏ —ñ –¥–æ –∫–æ–∂–Ω–æ—ó –∑ –Ω–∏—Ö –ø–æ—Ç—Ä—ñ–±–Ω–µ —Å–≤–æ—î –æ–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è.

–ù–∞–ø—Ä–∏–∫–ª–∞–¥, —Å–ø—Ä–æ–±–∞ –æ–±—á–∏—Å–ª–∏—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏—Ä–∞–∑—É $\displaystyle\frac{1}{\sqrt{x}}$ –º–æ–∂–µ –ø—Ä–∏–∑–≤–µ—Å—Ç–∏ –¥–æ –¥–≤–æ—Ö –≤–∏–¥—ñ–≤ –ø–æ–º–∏–ª–æ–∫: –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å (`ZeroDivisionError`) —ñ —Å–ø—Ä–æ–±–∞ –¥–æ–±—É—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∏–π –∫–æ—Ä—ñ–Ω—å –∑ –≤—ñ–¥'—î–º–Ω–æ–≥–æ —á–∏—Å–ª–∞ (`ValueError`). –í –æ–±–æ—Ö —Ü–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –ø–æ–≤–∏–Ω–µ–Ω –æ—Ç—Ä–∏–º–∞—Ç–∏ —Ä—ñ–∑–Ω—ñ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ –ø—Ä–∏—á–∏–Ω—É –∑–±–æ—é –ø—Ä–æ–≥—Ä–∞–º–∏.

–î–ª—è —Ç–æ–≥–æ, —â–æ–± —Ä—ñ–∑–Ω—ñ –≤–∏–Ω—è—Ç–∫–∏ –æ–±—Ä–æ–±–ª—é–≤–∞—Ç–∏ –ø–æ-—Ä—ñ–∑–Ω–æ–º—É, –ø—Ä–æ—Å—Ç–æ –¥–æ–¥–∞—é—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –±–ª–æ–∫—ñ–≤ `except`, –±—ñ–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑ –Ω–∏—Ö –∑–∞–∑–Ω–∞—á–∞—é—á–∏, –Ω–∞ —è–∫–∏–π —Å–∞–º–µ –≤–∏–Ω—è—Ç–æ–∫ —Ü–µ–π `except` –ø–æ–≤–∏–Ω–µ–Ω –≤—ñ–¥—Ä–µ–∞–≥—É–≤–∞—Ç–∏.

````python
import math
def division_sqrt(x):
    try:
        y = 1/math.sqrt(x)
        return y
    except ZeroDivisionError:
        return '–ü–û–ú–ò–õ–ö–ê! –°–ø—Ä–æ–±–∞ –ø–æ–¥—ñ–ª–∏—Ç–∏ –Ω–∞ –Ω—É–ª—å!'
    except ValueError:
        return "–ü–û–ú–ò–õ–ö–ê! –°–ø—Ä–æ–±–∞ –¥–æ–±—É—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∏–π –∫–æ—Ä—ñ–Ω—å –∑ –≤—ñ–¥'—î–º–Ω–æ–≥–æ —á–∏—Å–ª–∞!"
print(division_sqrt(4))
print(division_sqrt(-4))
print(division_sqrt(0))
````

–û–±—Ä–æ–±–∫—É –ø–æ–º–∏–ª–æ–∫ –º–æ–∂–Ω–∞ –∑–∞–ø—Ä–æ–≥—Ä–∞–º—É–≤–∞—Ç–∏ –π —ñ–Ω—à–∏–º —á–∏–Ω–æ–º, —è–∫—â–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —ñ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—é `raise` (*–∞–Ω–≥–ª.* "–ø—ñ–¥–Ω—ñ–º–∞—Ç–∏"). –¶–µ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–∏–π —ñ–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç, —è–∫–∏–π –¥–æ–∑–≤–æ–ª—è—î –ø—Ä–æ–≥—Ä–∞–º—ñ—Å—Ç—É —è–≤–Ω–æ –≤–∫–∞–∑—É–≤–∞—Ç–∏ –Ω–∞ –≤–∏–Ω–∏–∫–Ω–µ–Ω–Ω—è –ø–æ–º–∏–ª–æ–∫ —É –ø—Ä–æ–≥—Ä–∞–º—ñ.

–í —Ü—ñ–ª–æ–º—É, –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è `raise` –º–∞—î —Ç–∞–∫–∏–π –≤–∏–≥–ª—è–¥:

```python
raise Exception ('–ü–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ –ø–æ–º–∏–ª–∫—É')
```
–¥–µ **`Exception`** - —Ç–∏–ø –≤–∏–Ω—è—Ç–∫—É (*—Ç—É—Ç –º–æ–∂—É—Ç—å –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω—ñ —è–∫ –≤–±—É–¥–æ–≤–∞–Ω—ñ —Ç–∏–ø–∏ –≤–∏–Ω—è—Ç–∫—ñ–≤, —Ç–∞–∫ —ñ –≤–∏–∑–Ω–∞—á–µ–Ω—ñ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–µ–º*), –∞ **`–ü–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ –ø–æ–º–∏–ª–∫—É`** - —Ä—è–¥–æ–∫, —è–∫–∏–π —Ü—é –ø–æ–º–∏–ª–∫—É –æ–ø–∏—Å—É—î.

````python
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError ('–î—ñ–ª–∏—Ç–∏ –Ω–∞ –Ω—É–ª—å –Ω–µ –º–æ–∂–Ω–∞!')
    return a / b
try:
    print(divide(10, 2))
    print(divide(8, 0))
    print(divide(9, 4))
except ZeroDivisionError as error:
    print(f'–ü–æ–º–∏–ª–∫–∞: {error}')
````

–¢—É—Ç —É —Ä—è–¥–∫—É **3**: `raise ZeroDivisionError ('–î—ñ–ª–∏—Ç–∏ –Ω–∞ –Ω—É–ª—å –Ω–µ –º–æ–∂–Ω–∞!')` –º–∏, —è–∫ —Ç—ñ–ª—å–∫–∏ –≤–∏—è–≤–∏–ª–æ—Å—è, —â–æ –∑–º—ñ–Ω–Ω–∞ `b` –¥–æ—Ä—ñ–≤–Ω—é—î –Ω—É–ª—é, –æ–¥—Ä–∞–∑—É –∂ —à—Ç—É—á–Ω–æ —Å—Ç–≤–æ—Ä—é—î–º–æ –≤–∏–Ω—è—Ç–æ–∫, –Ω–µ –¥–æ–∂–∏–¥–∞—é—á–∏—Å—å, –ø–æ–∫–∏ —Ñ—É–Ω–∫—Ü—ñ—è `divide` –ø–æ—á–Ω–µ –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –π –Ω–∞—Ç–∫–Ω–µ—Ç—å—Å—è –Ω–∞ –ø–æ–º–∏–ª–∫—É. –¢–æ–±—Ç–æ, —è–∫ –±–∞—á–∏–º–æ, –ø–æ–¥—ñ—è –≤–∏–Ω—è—Ç–∫—É —â–µ –Ω–µ –Ω–∞—Å—Ç–∞–ª–∞ (*–¥—ñ–ª–µ–Ω–Ω—è —â–µ –Ω–µ –≤—ñ–¥–±—É–≤–∞–ª–æ—Å—å, –≤–æ–Ω–æ –±—É–¥–µ –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏—Å—è –≤ —Ä—è–¥–∫—É* **4**), –∞ –º–∏ –≤–∂–µ –∑–∞–∑–Ω–∞—á–∞—î–º–æ, —â–æ —Å–∏—Ç—É–∞—Ü—ñ—è, –∫–æ–ª–∏ –∑–º—ñ–Ω–Ω–∞ `b` –¥–æ—Ä—ñ–≤–Ω—é—î –Ω—É–ª—é, –ø–æ–≤–∏–Ω–Ω–∞ –±—É—Ç–∏ –æ–ø—Ä–∞—Ü—å–æ–≤–∞–Ω–∞ —è–∫ –≤–∏–Ω—è—Ç–æ–∫.

–¢–µ–ø–µ—Ä, –∫–æ–ª–∏ –ø—Ä–∏ —Å–ø—Ä–æ–±—ñ –≤–∏–∫–æ–Ω–∞—Ç–∏ —Ä—è–¥–æ–∫ **7** `print(divide(8, 0))` "–ø—ñ–¥–Ω—ñ–º–∞—î—Ç—å—Å—è" –≤–∏–Ω—è—Ç–æ–∫ `ZeroDivisionError`, —Ç–æ `except` —Ç–ª—É–º–∞—á–∏—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ –ø–æ–º–∏–ª–∫—É `'–î—ñ–ª–∏—Ç–∏ –Ω–∞ –Ω—É–ª—å –Ω–µ –º–æ–∂–Ω–∞!'`—è–∫ –∑–Ω–∞—á–µ–Ω–Ω—è –∑–º—ñ–Ω–Ω–æ—ó `error` —ñ –≤–∏–≤–æ–¥–∏—Ç—å –π–æ–≥–æ —É –ø—ñ–¥—Å—É–º–∫–æ–≤–æ–º—É –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—ñ `print(f'–ü–æ–º–∏–ª–∫–∞: {error}')` (—Ä—è–¥–æ–∫ **10**).

**–ó–≤–µ—Ä–Ω—ñ—Ç—å —É–≤–∞–≥—É:** –ø—ñ—Å–ª—è –Ω–∞—Å—Ç–∞–Ω–Ω—è –ø–æ–¥—ñ—ó –≤–∏–Ω—è—Ç–∫—É, —Ü—ñ–ª–∫–æ–º –Ω–æ—Ä–º–∞–ª—å–Ω–∏–π —Ä—è–¥–æ–∫ **8**: `print(divide(9, 4))`, –≤ —è–∫–æ–º—É –Ω–µ –º—ñ—Å—Ç–∏—Ç—å—Å—è –Ω—ñ—è–∫–∏—Ö –≤–∏–Ω—è—Ç–∫—ñ–≤, —É–∂–µ ***–Ω–µ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è***! –ù–∞—à–∞ –ø—Ä–æ–≥—Ä–∞–º–∞ –∑–∞–≤–µ—Ä—à–∏–ª–∞—Å—è –ø–æ `except`!

–û—Å—å —â–µ –æ–¥–∏–Ω –ø—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è `raise`:

````python
try:
    a = float(input('–ó–∞–¥–∞–π—Ç–µ —Å—Ç–æ—Ä–æ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–∞: '))
    if a < 0:
        raise ValueError("–°—Ç–æ—Ä–æ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–∞ –Ω–µ –º–æ–∂–µ –±—É—Ç–∏ –≤—ñ–¥'—î–º–Ω–æ—é!")
    print(f'–ü–ª–æ—â–∞ –∫–≤–∞–¥—Ä–∞—Ç–∞: {a**2}')
except ValueError as error:
    print(f'–£–≤–∞–∂–Ω—ñ—à–µ: {error}')
````

–ê –º–æ–∂–Ω–∞ –π —Ç–∞–∫:

````python
try:
    a = input('–ó–∞–¥–∞–π—Ç–µ —Å—Ç–æ—Ä–æ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–∞: ')
    if ',' in a:
        raise TypeError('–î–µ—Å—è—Ç–∫–æ–≤–∏–π –¥—Ä—ñ–± –∑–∞–ø–∏—Å—É–π—Ç–µ —á–µ—Ä–µ–∑ –∫—Ä–∞–ø–∫—É, –∞ –Ω–µ —á–µ—Ä–µ–∑ –∫–æ–º—É')
    else:
        print(f'–ü–ª–æ—â–∞ –∫–≤–∞–¥—Ä–∞—Ç–∞: {float(a)**2}')
except TypeError as error:
    print(f'–ü–æ–º–∏–ª–∫–∞: {error}')
````

## –Ø–∫ –¥—ñ–∑–Ω–∞—Ç–∏—Å—è —Ç–∏–ø –≤–∏–Ω—è—Ç–∫—É?

–ü—Ä–æ—Å—Ç–æ –Ω–∞–ø–∏—à—ñ—Ç—å –∫–æ–¥, –≤ —è–∫–æ–º—É –Ω–∞–≤–º–∏—Å–Ω–æ, –≤ —Ö–æ–¥—ñ —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è –∑—Ä–æ–±—ñ—Ç—å —è–∫—É—Å—å –Ω–µ–∫–æ—Ä–µ–∫—Ç–Ω—ñ—Å—Ç—å —ñ –∑–∞–ø—É—Å—Ç—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É –Ω–∞ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è. Python —É—Å–µ –≤–∞–º —Å–∞–º –Ω–∞–ø–∏—à–µ üôÉ

````python
print(2025**202.5)
````

–°–ø—Ä–æ–±–∞ –æ–±—á–∏—Å–ª–∏—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏—Ä–∞–∑—É $2025^{202,5}$ –≤–∏–∫–ª–∏–∫–∞–ª–∞ –≤–∏–Ω—è—Ç–æ–∫ `OverflowError` - —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–æ—ó –¥—ñ—ó –∑–∞–Ω–∞–¥—Ç–æ –≤–µ–ª–∏–∫–∏–π –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è —É —Ü—å–æ–º—É —Ç–∏–ø–æ–≤—ñ –¥–∞–Ω–∏—Ö.

–û—Ç–∂–µ, —Ç–µ–ø–µ—Ä –º–∏ –∑–Ω–∞—î–º–æ, —è–∫ –æ–±—Ä–æ–±–∏—Ç–∏ —Ü–µ–π –≤–∏–Ω—è—Ç–æ–∫ —É –ø—Ä–æ–≥—Ä–∞–º—ñ:

````python
def power(x, y):
    try:
        result = x ** y
        return result
    except OverflowError:
        return '–†–µ–∑—É–ª—å—Ç–∞—Ç –Ω–µ–º–æ–∂–ª–∏–≤–æ –∑–∞–ø–∏—Å–∞—Ç–∏.'
print(power(2025, 200))
print(power(2025, 202.5))
````

## –§—ñ–Ω–∞–ª—ñ–∑–∞—Ü—ñ—è `try`

–Ü–Ω–∫–æ–ª–∏ –≤–∏–Ω–∏–∫–∞—î –ø–æ—Ç—Ä–µ–±–∞ –ø—ñ—Å–ª—è –≤–∏–∫–æ–Ω–∞–Ω–Ω—è `try` –¥–æ–¥–∞—Ç–∏ —â–µ —â–æ—Å—å. –¢–æ–¥—ñ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å `finally` –∞–±–æ `else`.

–ü—ñ—Å–ª—è `finally` —Ä–æ–∑–º—ñ—â—É—î—Ç—å—Å—è –∫–æ–¥, —è–∫–∏–π –±—É–¥–µ –≤–∏–∫–æ–Ω–∞–Ω–æ ***–Ω–µ–∑–∞–ª–µ–∂–Ω–æ*** –≤—ñ–¥ —Ç–æ–≥–æ, –≤–∏–∫–æ–Ω–∞–≤—Å—è `try` –∑ –≤–∏–Ω—è—Ç–∫–æ–º, —á–∏ –±–µ–∑.

````python
def divide(x,y):
    try:
        result = x/y
        print(f'{x}/{y}={result}')
    except:
        print("–ü–æ–º–∏–ª–∫–∞: –¥—ñ–ª—å–Ω–∏–∫ –Ω–µ –º–æ–∂–µ –¥–æ—Ä—ñ–≤–Ω—é–≤–∞—Ç–∏ –Ω—É–ª—é!")
    finally:
        print("Game over!")
divide(10, 5)
divide(10, 0)
````

–£ –ø–µ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É –≤–∏–Ω—è—Ç–∫—É –Ω–µ —Å—Ç–∞–ª–æ—Å—è, –∞ —É –¥—Ä—É–≥–æ–º—É –≤–∏–Ω–∏–∫–ª–∞ –ø–æ–º–∏–ª–∫–∞. –í –æ–±–æ—Ö –≤–∏–ø–∞–¥–∫–∞—Ö, –Ω–µ–∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó `divide`, –±—É–ª–æ –≤–∏–≤–µ–¥–µ–Ω–µ –∑–∞–∫–ª—é—á–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è `Game over!`

–©–æ –≤–∏–≤–µ–¥–µ —Ü—è —Ñ—É–Ω–∫—Ü—ñ—è? üôÉ

````python
def func():
    try:
        return 'yes'
    finally:
        return 'no'
print(func())
````

–ê —á–æ–º—É —Ç–∞–∫?

–Ø–∫—â–æ "–∑–∞–∫–ª—é—á–Ω–∏–π –∫–æ–¥" –ø–æ–≤–∏–Ω–µ–Ω –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏—Å—è —Ç—ñ–ª—å–∫–∏ –≤ —Ç–æ–º—É –≤–∏–ø–∞–¥–∫—É, —è–∫—â–æ –≤–∏–Ω—è—Ç–∫—É ***–Ω–µ —Å—Ç–∞–ª–æ—Å—å***, —Ç–æ–¥—ñ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è `else`.

````python
def remainder(x):
    try:
        num = x % 2
    except TypeError:
        print('–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∏–π —Ç–∏–ø –¥–∞–Ω–∏—Ö! –ó–∞–¥–∞–π—Ç–µ —Ü—ñ–ª–µ —á–∏—Å–ª–æ!')
    else:
        print(f'–û—Å—Ç–∞—á–∞: {num}')
a = 83
b = 10,2
remainder(a)
remainder(b)
````

## –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–ª–∞—Å–Ω–∏—Ö –≤–∏–Ω—è—Ç–∫—ñ–≤

–û–∫—Ä—ñ–º –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤–±—É–¥–æ–≤–∞–Ω–∏—Ö —Ç–∏–ø—ñ–≤ –≤–∏–Ω—è—Ç–∫—ñ–≤, –º–æ–∂–Ω–∞ —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –π –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `raise` "–ø—ñ–¥–Ω—ñ–º–∞—Ç–∏" —Å–≤–æ—ó –≤–ª–∞—Å–Ω—ñ –≤–∏–Ω—è—Ç–∫–∏, —è–∫—ñ —Ç–∞–∫ —Å–∞–º–æ –æ–±—Ä–æ–±–ª—è—Ç–∏–º—É—Ç—å—Å—è –ø—Ä–æ–≥—Ä–∞–º–æ—é.

–ü–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å —Ä–æ–±–æ—Ç–∏ –∑ –≤–ª–∞—Å–Ω–∏–º–∏ –≤–∏–Ω—è—Ç–∫–∞–º–∏ —Ç–∞–∫–∞:
+ –°—Ç–≤–æ—Ä—é—î–º–æ —Å–≤—ñ–π –∫–ª–∞—Å –≤–∏–Ω—è—Ç–∫—É, —è–∫ –¥–æ—á—ñ—Ä–Ω—ñ–π –¥–æ –∫–ª–∞—Å—É –≤–±—É–¥–æ–≤–∞–Ω–∏—Ö –≤–∏–Ω—è—Ç–∫—ñ–≤ `Exception`.
+ –û–ø–∏—Å—É—î–º–æ –π–æ–≥–æ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –π –º–µ—Ç–æ–¥–∏ (*—è–∫—â–æ –ø–æ—Ç—Ä—ñ–±–Ω–æ*).
+ –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ —Ç–∞–∫ —Å–∞–º–æ, —è–∫ —ñ –±—É–¥—å-—è–∫–∏–π –∑ –≤–±—É–¥–æ–≤–∞–Ω–∏—Ö –≤–∏–Ω—è—Ç–∫—ñ–≤.

````python
class AgeError(Exception): # –°—Ç–≤–æ—Ä—é—î–º–æ –≤–ª–∞—Å–Ω–∏–π –∫–ª–∞—Å –≤–∏–Ω—è—Ç–∫—ñ–≤
    pass # –ù—ñ—è–∫—ñ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –π –º–µ—Ç–æ–¥–∏ –Ω–∞–º –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω—ñ, —Ç–æ–º—É pass

def registration_check(age):
    if age < 18:
        raise AgeError('–í—Ö—ñ–¥ —Ç—ñ–ª—å–∫–∏ –¥–ª—è –ø–æ–≤–Ω–æ–ª—ñ—Ç–Ω—ñ—Ö!') # "–ü—ñ–¥–Ω—ñ–º–∞—î–º–æ" —Å–≤—ñ–π –≤–∏–Ω—è—Ç–æ–∫
    else:
        print('–ö–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –∑–∞—Ä–µ—î—Å—Ç—Ä–æ–≤–∞–Ω–æ')

try:
    age = int(input('–í–∞—à –≤—ñ–∫: '))
    registration_check(age)
except AgeError as error:
    print(error)
````

–°—Ç–≤–æ—Ä–∏–º–æ —Å–≤—ñ–π –≤–∏–Ω—è—Ç–æ–∫ –¥–ª—è –≤—ñ–¥–∑–Ω–∞—á–µ–Ω–Ω—è –≤—ñ–¥—Å—É—Ç–Ω–æ—Å—Ç—ñ –Ω–∞ —Å–∫–ª–∞–¥—ñ —Ç–æ–≤–∞—Ä—ñ–≤ –ø–µ–≤–Ω–æ–≥–æ –≤–∏–¥—É.

````python
class StorageError(Exception):
    def __init__(self, item_name):
        self.item_name = item_name #–Ω–∞–∑–≤–∞ —Ç–æ–≤–∞—Ä—É
        super().__init__(f"–¢–æ–≤–∞—Ä \'{self.item_name}\' –≤—ñ–¥—Å—É—Ç–Ω—ñ–π –≤ –ø–æ—Ç—Ä—ñ–±–Ω—ñ–π –∫—ñ–ª—å–∫–æ—Å—Ç—ñ")

def storage_order(item, number, stock): #–Ω–∞–∑–≤–∞, –∫—ñ–ª—å–∫—ñ—Å—Ç—å, —Å–∫–ª–∞–¥
    if item not in stock or stock[item] < number:
        raise StorageError(item)
    else:
        stock[item] -= number
        print(f"–í–∏–¥–∞–Ω–æ {number} –æ–¥–∏–Ω–∏—Ü—å —Ç–æ–≤–∞—Ä—É \'{item}\'")
        print(f'–ó–∞–ª–∏—à–æ–∫: {stock[item]} –æ–¥–∏–Ω–∏—Ü—å')

storage_stock = {'—è–±–ª—É–∫–∞': 50, '–±–∞–Ω–∞–Ω–∏': 30, '–∞–Ω–∞–Ω–∞—Å–∏': 100} #–∑–∞–ø–∞—Å –Ω–∞ —Å–∫–ª–∞–¥—ñ

try:
    storage_order('—è–±–ª—É–∫–∞', 20, storage_stock)
    storage_order('–±–∞–Ω–∞–Ω–∏', 50, storage_stock)
except StorageError as error:
    print(error)
````
